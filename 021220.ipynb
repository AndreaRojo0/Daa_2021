```
def fnRec(x):
  if x==0:
    print("Stop")
  else:
    fnRec(x-1)
    print(x)

def main():
  print("inicio del programa")
  fnRec(5)
  print("final del programa")

main()
```

```
def printRev(x):
  if x>0:
    print(x)
    printRev(x-1)

printRev(3)
```

```
def fibonacci(n):
    if n < 2:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
fibonacci(8)
```
